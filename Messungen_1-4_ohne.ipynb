```python
import matplotlib.pyplot as plt
import numpy as np
import lmfit
import os 
```

```python
for i in os.listdir('Data'):
    path = 'Data/'+i
    if i == 'SI_111_30_15.xy' or i == 'si111_30_20_ohne.xy' or i == 'si111_40_20_ohne_ni.xy' or i == 'Si_111_40_40_ohne_Fi.xy':
        with open(path) as tf:
            data = np.loadtxt(tf,delimiter=" ").T
        fig = plt.figure(figsize=(10,9), dpi=80,linewidth=50)
        plt.plot(data[0],data[1],label='Data')
        #plt.ylim(-1000,30000)
        plt.title(label=i.split('.')[0])
        plt.legend()
        plt.show()
```
