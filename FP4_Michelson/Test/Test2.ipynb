```python
import numpy as np
import matplotlib.pyplot as plt

# Set the sampling frequency and time interval
fs = 1000  # Hz
T = 1/fs  # seconds

# Generate a random frequency between 5 and 50 Hz
f = 12#np.random.uniform(5, 50)

# Generate a time vector
t = np.arange(0, 1, T)

# Generate a sinusoidal signal with the random frequency
x = np.sin(2*np.pi*f*t)



# Compute the FFT of the signal
X = np.fft.fft(x)

# Compute the frequency vector for the FFT output
freqs = np.fft.fftfreq(len(X), d=T)

# Find the peak frequency and its magnitude
peak_idx = np.argmax(np.abs(X))
peak_freq = freqs[peak_idx]
peak_mag = np.abs(X[peak_idx])

# Print the peak frequency and its magnitude
print(f"Peak frequency: {peak_freq} Hz")
print(f"Peak magnitude: {peak_mag}")

# Plot the signal and its FFT
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.plot(t, x)
ax1.set_ylabel("Amplitude")
ax2.plot(freqs, np.abs(X))
ax2.set_ylabel("FFT magnitude")
ax2.set_xlabel("Frequency (Hz)")
plt.xlim(0,20)
plt.show()
```
