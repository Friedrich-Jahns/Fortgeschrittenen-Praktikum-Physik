```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fft as fft
```

```python
sin = []
for i in np.arange(0,100,0.001):
    sin.append(np.sin(0.5*i)+np.sin(5*i))
```

```python
plt.plot(np.arange(0,100,0.001),sin)
```

```python
import numpy as np

# Load interferogram data from a file or generate it directly
data = [np.arange(0,100,0.001),sin]


# Extract the position and intensity values from the data
position = data[0]  # units of length (e.g. meters)
intensity = data[1]

# Define the sampling rate of the interferogram
sampling_rate = 1 / np.mean(np.diff(position))  # samples per unit length (e.g. samples per meter)

# Perform FFT on the intensity values of the interferogram
fft_data = np.fft.fft(intensity)

# Calculate the frequencies represented in the FFT data
freqs = np.fft.fftfreq(len(fft_data), d=1/sampling_rate)

# Calculate the corresponding wavenumbers and wavelengths
wavenumbers = 2 * np.pi * freqs
wavelengths = 1 / wavenumbers

# Plot the FFT data (optional)
import matplotlib.pyplot as plt
plt.plot(wavelengths, np.abs(fft_data))
#plt.vlines(3.39e-6,0,10000,colors="r")

#plt.xlim(0,0.5)
plt.xlabel("Wavelength (units of length)")
plt.ylabel("FFT magnitude")
plt.show()
```
