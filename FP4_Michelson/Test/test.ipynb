```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fft 
```

```python
sin = []
for i in np.arange(0,10,0.001):
    sin.append(np.sin(i*2)+np.sin(40*i))
```

```python
plt.plot(np.arange(0,10,0.001),sin)
```

```python
fftsin = fft.rfft(x=sin, axis=0)
fftsin = np.absolute(fftsin)
```

```python
plt.plot(np.arange(0,10,0.001),fftsin)
```
