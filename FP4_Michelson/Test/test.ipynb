```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fft as fft
```

```python
sin = []
for i in np.arange(0,10,0.001):
    sin.append(np.sin(np.rad2deg(0.5*i))+np.sin(np.rad2deg(5*i)))
```

```python
plt.plot(np.arange(0,10,0.001),sin)
```

```python


N=len(np.arange(0,10,0.001))
sr = 1 / 0.001
T = N/sr

yf = fft.rfft(sin)
xf = fft.rfftfreq(N,T)
```

```python
plt.plot(xf,np.abs(yf))
plt.xlim(0,50)
```
