```python
with open("Results/amplitudes.txt", mode='r') as tf:
    data = np.genfromtxt(tf,delimiter="\t",skip_header=1).T



pressure = data[0]
amp1 = data[1]
amperr1 = data[2]
amp2 = data[3]
amperr2 = data[4]
amp3 = data[5]
amperr3 = data[6]
amp4 = data[7]
amperr4 = data[8]

fig = plt.figure(figsize=(8,6),dpi=80,linewidth=50)

plt.errorbar(x=pressure,y=amp1,yerr=amperr1,label=r'$^{226}$Ra',fmt='--o',ecolor='gray', elinewidth=3, capsize=0,zorder=5)
plt.errorbar(x=pressure,y=amp2,yerr=amperr2,label=r'$^{222}$Rn/$^{210}$Po',fmt='--o',ecolor='gray', elinewidth=3, capsize=0,zorder=5)
plt.errorbar(x=pressure,y=amp3,yerr=amperr3,label=r'$^{218}$Po',fmt='--o',ecolor='gray', elinewidth=3, capsize=0,zorder=5)
plt.errorbar(x=pressure,y=amp4,yerr=amperr4,label=r'$^{214}$Po',fmt='--o',ecolor='gray', elinewidth=3, capsize=0,zorder=5)


plt.grid()
plt.legend(fontsize = 15)
plt.xlabel('Pressure [Torr]',fontsize = 20)
plt.ylabel('Counts',fontsize = 20)
plt.savefig('Results/ConsOfCountrate.pdf')
plt.show()
```
