```python
from uncertainties import unumpy as up
from uncertainties import ufloat 
import numpy as np
```

```python
def ksize(fwhm,theta,l = 1.5425,k=1):
    return  (k*l)/(up.radians(fwhm)*up.cos(up.radians(theta/2)))
```

```python
with open('results_GeOx/GeOx.txt') as tf:
        data = np.loadtxt(tf,delimiter="\t",skiprows=1,dtype="str")
fwhm1 = ufloat(data[1,0],float(data[1,1]))
fwhm2 = ufloat(data[1,2],float(data[1,3]))
fwhm3 = ufloat(data[1,4],float(data[1,5]))
fwhm4 = ufloat(data[1,6],float(data[1,7]))
fwhm5 = ufloat(data[1,8],float(data[1,9]))
fwhm6 = ufloat(data[1,10],float(data[1,11]))
fwhm7 = ufloat(data[1,12],float(data[1,13]))
fwhm8 = ufloat(data[1,14],float(data[1,15]))
fwhm9 = ufloat(data[1,16],float(data[1,17]))
fwhm10 = ufloat(data[1,18],float(data[1,19]))
fwhm11 = ufloat(data[1,20],float(data[1,21]))
fwhm12 = ufloat(data[1,22],float(data[1,23]))
fwhm13 = ufloat(data[1,24],float(data[1,25]))
fwhm14 = ufloat(data[1,26],float(data[1,27]))

mu1 = ufloat(data[0,0],float(data[0,1]))
mu2 = ufloat(data[0,2],float(data[0,3]))
mu3 = ufloat(data[0,4],float(data[0,5]))
mu4 = ufloat(data[0,6],float(data[0,7]))
mu5 = ufloat(data[0,8],float(data[0,9]))
mu6 = ufloat(data[0,10],float(data[0,11]))
mu7 = ufloat(data[0,12],float(data[0,13]))
mu8 = ufloat(data[0,14],float(data[0,15]))
mu9 = ufloat(data[0,16],float(data[0,17]))
mu10 = ufloat(data[0,18],float(data[0,19]))
mu11 = ufloat(data[0,20],float(data[0,21]))
mu12 = ufloat(data[0,22],float(data[0,23]))
mu13 = ufloat(data[0,24],float(data[0,25]))
mu14 = ufloat(data[0,26],float(data[0,27]))
```
