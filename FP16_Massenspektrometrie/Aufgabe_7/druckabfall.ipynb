```python
def pos(x):
    for i in range(len(data[0])):
        if data[1,i]>=x:
            return i
            break
```

```python
yscale = 1e8
xscale = 1/1000
with open('Data/plot.csv', mode='r') as tf:
        data = np.genfromtxt((line.replace(',', '.') for line in tf),delimiter=";",skip_header=25).T

von = pos(35000)
bis = pos(60000)




fig = plt.figure(figsize=(20,7),dpi=80,linewidth=50)
plt.plot(xscale*data[1,von:bis],yscale*data[2,von:bis],linewidth=1.5,label=r'$H_2O$')
plt.plot(xscale*data[1,von:bis],yscale*data[3,von:bis],linewidth=1.5,label=r'$N_2$')
plt.plot(xscale*data[1,von:bis],yscale*data[4,von:bis],linewidth=1.5,label=r'$O_2$')
plt.plot(xscale*data[1,von:bis],yscale*data[5,von:bis],linewidth=1.5,label=r'$Ar$')
plt.plot(xscale*data[1,von:bis],yscale*data[6,von:bis],linewidth=1.5,label=r'$CO_2$')
plt.legend(fontsize=15)

plt.grid()
plt.xlabel('Zeit [s]')
plt.ylabel('Partialdruck [Torr $10^{'+str(1/yscale).split("e")[1]+'}$]')
plt.legend()
plt.savefig('Results/druckabfall.pdf')
```
